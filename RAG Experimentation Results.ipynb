# 🔬 RAG Experimentation Results

This document summarizes the comparative analysis of different Retrieval-Augmented Generation (RAG) strategies used for the Academic Study Assistant.

---

## 📊 Summary of Experiments

| Experiment | Strategy | Chunk Size | Overlap | Key Finding |
| :--- | :--- | :--- | :--- | :--- |
| **01** | Baseline | 1000 | 0 | Fast but loses context in technical lists. |
| **02** | Fixed Overlap | 1000 | 200 | Improved continuity between paragraphs. |
| **03** | Recursive | Variable | 200 | **Best Performer:** Handles nested OS concepts best. |

---

## 🧪 Detailed Findings

### Experiment 1: The Baseline (Standard RAG)
* **Method:** Large chunks with no overlap.
* **Result:** The model frequently failed on "List all steps of..." questions because the steps were split across two different database entries.
* **Accuracy:** 65%

### Experiment 2: Fixed Overlap RAG
* **Method:** Added a 200-character buffer between chunks.
* **Result:** Significant improvement in "Contextual Retrieval." Technical terms that appear at the end of a page were no longer cut off.
* **Accuracy:** 82%

### Experiment 3: Recursive Character Splitting (Experimental)
* **Method:** Used `RecursiveCharacterTextSplitter` to split by paragraphs first, then sentences.
* **Result:** This provided the most cohesive "Study Notes." TinyLlama produced more coherent answers because the retrieved context was semantically complete.
* **Accuracy:** 91%
